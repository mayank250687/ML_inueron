```
fig , ax =plt.subplots(2,2,figsize=(10,10))

x= np.arange(1,5)
ax[0][0].plot(x,x*x,color='r')
ax[0][0].set_title('square')
ax[0][1].plot(x,np.sqrt(x), color='g')
ax[0][1].set_title('square root')
ax[1][0].plot(x,np.exp(x),color='b')
ax[1][0].set_title('Exp')
ax[1][1].plot(x,np.log10(x),color='k')
ax[1][1].set_title('log')
plt.show()

```

```
fig , ax =plt.subplots(2,2,figsize=(10,10))
x= np.arange(1,5)
ax[0][0].plot(x,x*x,color='r')
ax[0][0].grid(True)
ax[0][0].set_title('square')
ax[0][1].plot(x,np.sqrt(x), color='g')
ax[0][1].grid(color='g', ls='-.',lw=0.35)
ax[0][1].set_title('square root')
ax[1][0].plot(x,np.exp(x),color='b')
ax[1][0].set_title('Exp')
ax[1][1].plot(x,np.log10(x),color='k')
ax[1][1].set_title('log')
plt.show()
```

```
water=[10,20,50,60,40]
oil=[90,80,50,40,60]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=np.arange(5)
ax.bar(x,water,width=0.35,color='lightblue')
ax.bar(x,oil,width=0.35,bottom=water,color='y')
ax.set_xlabel('Indicators')
ax.set_ylabel('values')
ax.legend(labels=['Water','Oil'])
ax.set_title('Water -Oil  Ration')

```

```
# hist() - histogram
# bar, step, stepfilled, barstacked
import matplotlib.pyplot as plt
import numpy as np
fig, ax =plt.subplots(1,1)
marks=np.array([20,60,25,30,19,35,38,45,50,60,81,89,70,78,90,91,95,61,66,53])
range=[0,20,40,60,80,100]
ax.hist(marks,bins=range,histtype='bar',color=['c'])
ax.set_title('histogram of results')
ax.set_xlabel('marks')
ax.set_ylabel('no of students')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
fig, ax =plt.subplots(1,1)
marks=np.array([20,60,25,30,19,35,38,45,50,60,81,89,70,78,90,91,95,61,66,53])

ax.hist(marks,histtype='step',range=(20,95))
ax.set_title('histogram of results')
ax.set_xlabel('marks')
ax.set_xticks([20,40,80,95])
ax.set_yticks([0,2,3,4])
ax.set_ylabel('no of students')
plt.show()
```

```
# twin axes
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
x=np.arange(1,11)
ax1.plot(x,np.log10(x),color='g',marker='s')
ax1.set_ylabel('log')
ax1.set_xlim(-10 ,10)

ax2=ax1.twinx()
ax2.plot(x,np.sqrt(x),color='r')
ax2.set_ylabel('square root')
fig.legend(labels=('log','sqrt'))
plt.show()


```
